## **Min Max scaling on the feature set (Take â€˜Survivedâ€™ as target)**

```python
df.dtypes
```

```python
# Dropping 'Name' and 'Ticket' as it contains no unique values

df.drop('Name', inplace=True, axis=1)
df.drop('Ticket', inplace=True, axis=1)
df
```

```python
# One hot encoding

df = pd.get_dummies(df)
df
```

```python
# Splitting into independent and dependent variable

y= df['Survived']
x = df.drop(['Survived','Fare'],axis=1)

# Splitting

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, random_state=42, test_size = 0.25)

# Linear model

from sklearn import linear_model
lr = linear_model.LinearRegression()
model = lr.fit(x_train, y_train)
predictions = model.predict(x_test)

# Mean squared error

from sklearn.metrics import mean_squared_error
print("MSE is: ", mean_squared_error(y_test,predictions))
from sklearn.metrics import r2_score
print("R-squared value is: ",r2_score(y_test,predictions))
```

```python
df = pd.DataFrame(data={'Predicted value':predictions,'Actual Value':y_test})
df
```

```python
# min max scaling

from sklearn import preprocessing
min_max=preprocessing.MinMaxScaler(feature_range=(0,1))
x=min_max.fit_transform(x)
x=pd.DataFrame(x)
x.describe()
```

```python
# After min max scaling

x_train,x_test,y_train,y_test = train_test_split(x, y, random_state=42, test_size = 0.25)
lr = linear_model.LinearRegression()
model = lr.fit(x_train, y_train)
predictions = model.predict(x_test)
print("MSE is: ", mean_squared_error(y_test,predictions))
print("R-squared value is: ",r2_score(y_test,predictions))
```

```python
df = pd.DataFrame(data={'Predicted value':predictions,'Actual Value':y_test})
df
```

# Conclusion : **We can see that the MSE and R-Squared value before and after  using Min Max Scaling method are the same**
