# Libraries

```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read File

```python
#Reading excel file
data=pd.read_excel('iris.xls')
```

# Checking the data

```python
#Checking the first 5 rows on the data
data.head()
```

```python
#Checking for any null values in the data
data.isnull().sum()
```

```python
#Checking the data types in each column
data.dtypes
```

```python
#Count of each species
plt.figure(figsize=(8,8))
plt.title('Count of species')
sns.countplot(x='Classification',data=data)
plt.show()
```

Insight : The data is well balanced as the data has 50 obeservations 

# Mean of each feature in the dataset

```python
#Calculating the mean of each column
data.mean(numeric_only=True,axis=0)
```

# Sepal length and Sepal width of each species

```python
#Using a scatterplot to compare between various species based on sepal length and sepal width
plt.figure(figsize=(12,8))
plt.title('Comparison between various species based on sepal length and width')
sns.scatterplot(x='SL',y='SW',hue='Classification',data=data)
plt.show()
```

Insights : 
 <br> - Iris-setosa has shorter sepal length but wider sepal.
 <br> - Iris-versicolor has average or balanced length as well as width.
 <br> - Iris-virginica has longer sepal lengths and shorter sepal widths.

# Petal length and width of each species

```python
#Using a lineplot to analyse the petal length and width of each species
plt.figure(figsize=(12,8))
plt.title('Comparison between various species based on sepal length and width')
sns.lineplot(x = "PW", y ="PL",hue = "Classification", data = data)
plt.show()
```

Insights :
<br>   -  Iris-setosa has the shortest petal length and petal width.
<br>   -  Iris-versicolor has average or balanced petal length and petal width.
<br>   -  Iris-virginica has the longest petal length and petal width.

# Correlation

```python
#Using a heatmap to analyse correlation between different attributes in the dataset
plt.figure(figsize=(10,8))
plt.title('Correlation Heatmap')
sns.heatmap(data.corr(),annot=True)
plt.show()
```

Insight : There is a correlation between sepal length and petal width as well as petal length and sepal length.
