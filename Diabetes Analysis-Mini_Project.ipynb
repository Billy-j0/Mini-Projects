# Libraries

```python
#Importing all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read CSV File

```python
#Reading and Storing the data 
data=pd.read_csv('diabetes_model.csv')
```

# Checking the data 

```python
#Checking the first 5 rows of the data
data.head()
```

```python
#Checking the data for any null values
data.isnull().sum()
```

# Countplot For Diabetes Outcome

```python
#Using countplot for checking the outcome of diabetes
plt.title('Countplot of Outcome')
fig=sns.countplot(x = data['Outcome'])
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.35, p.get_height()-150))
```

# Analysing the affect of Age on Outcome

```python
#Using box plot to analyse the affect of age on outcome
plt.figure(figsize=(8,8))
plt.title('Age v/s Outcome')
sns.boxplot(x='Outcome',y='Age',data=data)
```

Insight : It is observed that chances of being diabetic increases with age.

# Diabetes Pedigree Function

```python
#Using boxplot  to check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function
plt.figure(figsize=(8,8))
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
```

Insight : It is observed that there is a higher chance of being diabetic if it runs in the family.

#### Note : DiabetesPedigreeFunction: Diabetes pedigree function (a function which scores likelihood of diabetes based on family history) Age: Age (years) Outcome: Class variable (0 if non-diabetic, 1 if diabetic)

# Relationship between Insulin & Glucose with respect to Outcome

```python
#Using a scateerplot to check for the relationship with Insulin & Glucose and group the datapoints by Outcome
plt.figure(figsize=(8,8))
sns.scatterplot(x=data['Insulin'],y=data['Glucose'],hue=data['Outcome'])
```

Insight : There exists a linear relationship between insulin and glucose.

#  Correlation of features in the data

```python
#Using a heatmap to show the correlation of features in the data
plt.figure(figsize=(12,10)) 
sns.heatmap(data.corr(),annot=True)
```
